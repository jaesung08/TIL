# 문제4. 상품과 옵션 정보들을 담고 있는 새로운 객체를 만들어 반환하시오.
- [힌트] 상품 리스트와 옵션 리스트를 금융상품 코드를 기준으로 매칭할 수 있습니다.
- [힌트] 아래와 같은 순서로 데이터를 출력하며 진행합니다.
- 1. 응답을 json 형식으로 변환합니다.
- 2. key 값이 "result" 인 데이터를 변수에 저장합니다.
- 3. 2번의 결과 중 key 값이 "baseList" 인 데이터를 변수에 저장합니다.
- 4. 2번의 결과 중 key 값이 "optionList" 인 데이터를 변수에 저장합니다.
- 5. 3번에서 저장된 변수를 순회하며, 4번에서 저장된 값들에서 금융 상품 코드가 
-     같은 모든 데이터들을 가져와 새로운 딕셔너리로 저장합니다.
-     저장 시, 명세서에 맞게 출력되도록 저장합니다.
- 6. 5번에서 만든 딕셔너리를 결과 리스트에 추가합니다.

```python
def rename_dict_key(data_list, old_name, new_name):
    new_list = []
    for data in data_list:
        if old_name in data:
            data[new_name] = data.pop(old_name)
        new_list.append(data)
    return new_list

old_names = ['dcls_month', 'fin_co_no', 'fin_prdt_cd','intr_rate', 'intr_rate2','intr_rate_type', 'intr_rate_type_nm','save_trm' ]
new_names = ['공시 제출월', '금융회사 코드',' 금융상품 코드', '저축금리', '최고 우대금리','저축금리 유형', '저축금리 유형명','저축 기간']


def get_deposit_products():
    # 본인의 API KEY 로 수정합니다.
    api_key = "d8ac7b184792adf08a6f6bb6bd619d37"
    url = f'http://finlife.fss.or.kr/finlifeapi/depositProductsSearch.json?auth={api_key}&topFinGrpNo=020000&pageNo=1'
    
    result = requests.get(url).json().get('result')
    base_list = result.get('baseList')
    option_list = result.get("optionList")
    for i in range(0,8) :
        option_list_k = rename_dict_key(option_list, old_names[i] , new_names[i])
    
    # base_list에 속하는 회사의 이름과 상품명
    company_product_dict = {}
    for base in base_list:
        company_name = base['kor_co_nm']
        product_name = base['fin_prdt_nm']
        company_product_dict[base['fin_co_no']] = {'금융회사명': company_name, '금융상품명': product_name}

    # base_list의 fin_co_no과 같은 값을 가진 option_list들을 한 곳에 추출
    fin_co_no_set = set(item['fin_co_no'] for item in base_list)
    new_option_list = []
    for option in option_list_k:
        if option['금융회사 코드'] in fin_co_no_set:
            new_option_list.append(option)
    

    # 결과를 담을 리스트 생성
    result_list = []
    for option in new_option_list:
        fin_co_no = option['금융회사 코드']
        product_name = company_product_dict[fin_co_no]['금융상품명']
        company_name = company_product_dict[fin_co_no]['금융회사명']
        result_list.append({'금리정보': option, '금융상품명': product_name, '금융회사명': company_name})
        
        
    return result_list
```

```python
# 아래 코드는 수정하지 않습니다.
if __name__ == '__main__':
    # json 형태의 데이터 반환
    result = get_deposit_products()
    # prrint.prrint(): json 을 보기 좋은 형식으로 출력
    pprint.pprint(result)
```
